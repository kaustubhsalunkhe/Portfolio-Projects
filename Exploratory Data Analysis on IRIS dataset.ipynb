# Exploratory Data Analysis on IRIS dataset

```python
#important required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
#load the dataset
iris_data=load_iris()
```

```python
#Understand the dataset features and target variable
print("\n Features:\n",iris_data.feature_names)
print("\n Outcome Variables:\n",iris_data.target_names)
```

```python
# print the first few lines of data
#convert the dataset into  pandas dataframe
iris_df=pd.DataFrame(data=np.c_[iris_data['data'], iris_data['target']],
                     columns=iris_data['feature_names'] + ['species'])
```

```python
#Display the Dataset Header
iris_df.head()
```

```python
#Data Preprocessing
#Data Cleaning
print("\n Check Missing Values in Dataframe")
print(iris_df.isnull().sum())
```

```python
#Explore The Data
#Visualize Feature Distributions
plt.figure(figsize=(12, 6))
for i, feature in enumerate(iris_data.feature_names):
    plt.subplot(2, 2, i+1)
    sns.histplot(data=iris_df, x=feature, kde=True)
    plt.title(f"Distribution of {feature}")
plt.tight_layout()
plt.show()
```

```python
#Summary Statistics
print("\n Summary Statistics:\n")
print(iris_df.describe(include='all'))
```

```python
#Pairwise Relationships between Features
sns.pairplot(iris_df, hue='species', diag_kind='kde')
plt.suptitle("Pairwise Relationship Between Features", y=1.02) #(KDE) Kernel Density Estimate
plt.show()
```

```python
#Correlation Matrix And Heatmap
correlation_matrix = iris_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
#Class Ditribution Analysis
plt.figure(figsize=(8, 6))
sns.countplot(data=iris_df, x='species')
plt.title("Class Distribution")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```
